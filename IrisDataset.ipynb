```python
import pandas as pd
import numpy as np
import matplotlib .pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("iris.csv")
```

```python
del df["species"]
```

```python
a = KMeans(n_clusters=5)
a.fit(df)
```

```python
plt.scatter(df.sepal_length,df.petal_length,c=a.labels_)
plt.show()
```

```python
plt.scatter(df.sepal_width,df.petal_width,c=a.labels_)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
linkage_data = linkage(df, method='single', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
linkage_data2 = linkage(df, method='complete', metric='euclidean')
dendrogram(linkage_data2)

plt.show()
```
